```python
# Solusi integral menggunakan Metode Monte Carlo

import numpy as np
import random
import matplotlib.pyplot as plt

class IntegralMonteCarlo:
  def __init__(self, func, a, b, n):
    self.func = func # Fungsi yang akan diintegralkan
    self.a = a # Batas bawah integral
    self.b = b # Batas atas integral
    self.n = n # Banyak iteras

  def hitung_integral(self):
    x_ran = np.random.uniform(self.a, self.b, self.n) # Angka acak antara a dan b
    y_ran = np.random.uniform(0, self.func(self.b), self.n) # Angka acak antara 0 dan nilai fungsi b

    y_hasil = self.func(x_ran) # Nilai fungsi dari x_ran

    # Menentukan titik berada di bawah atau di atas y_hasil
    titik_dalam = y_ran <= y_hasil
    titik_luar = y_ran > y_hasil

    area_dalam = (self.b - self.a) * max(y_hasil) # Estimasi area dari fungsi
    integral = area_dalam * np.sum(titik_dalam) # Menghitung integral

    return x_ran, y_ran, integral, titik_dalam, titik_luar

  def visualisasi(self):
    x = np.linspace(self.a, self.b, n)
    y = self.func(x)

    x_ran, y_ran, integral, titik_dalam, titik_luar = self.hitung_integral()

    plt.plot(x, y, label='Fungsi f(x)', color = 'yellow', linewidth = 3)
    plt.scatter(x_ran[titik_dalam], y_ran[titik_dalam], color='red', label='Titik dalam area')
    plt.scatter(x_ran[titik_luar], y_ran[titik_luar], color='blue', label='Titik luar area')
    plt.xlabel('x')
    plt.ylabel('f(x) = y')
    plt.title('Hasil Integral Menggunakan Metode Monte Carlo')
    plt.legend()
    plt.show()

    return integral

def fungsi(x):
  return x**2
  # return x**2 + 4*x*np.sin(x)

a = int(input('Masukkan nilai a (batas bawah integral) : '))
b = int(input('Masukkan nilai b (batas atas integral) : '))
n = int(input('Masukkan banyak iterasi : '))

proses = IntegralMonteCarlo(fungsi, a, b, n)
integral = proses.hitung_integral()
hasil = proses.visualisasi()

print('Nilai integral adalah', hasil)
```
